# 1. Backpropagation in Detail
- squared error function:
$$e_i=\frac{1}{2}(t_i-o_i)^2$$

```python
# example who:
import numpy as np


who = np.array([[0.6, 0.5, 0.06],
                [0.3, 0.2, 0.8],
                [0.2, 0.1, 0.7],
                [0.12, 0.7, 0.54]])

e = np.array([0.1, 0.2, 0.3])

who * e
```

- 4개의 입력노드, 3개의 출력노드가 있다면 ⇒ 전체 error는 각 error와 각 가중치의 식으로 표현할 수 있음 <br>
$$e_{who}=who^T \cdot e$$
image.png

- derivation of error $E$ at wieght $s_{kj}$: 가중치 1개 값이 변할 때 전체 에러 즉, 비용함수는 얼마나 변하는지
$$E=\sum_{i=1}^n \frac{1}{2}(t_i-o_i)^2$$
$$\frac{\partial E}{\partial w_{kj}}=\frac{\partial}{\partial w_{kj}}\sum_{i=1}^n \frac{1}{2}(t_i-o_i)^2$$
- 비용함수는 가중치들의 합성함수이므로 chain rule 적용 가능
$$\frac{\partial E}{\partial w_{kj}}=\frac{\partial E}{\partial o_{k}}\cdot \frac{\partial o_k}{\partial w_{kj}}$$
- $o_k$ 출력노드 출력값은 sigmoid 함수를 통과하기 때문에 sigmoid 함수 미분 공식을 활용하면, 
  - sigmoide function derivative: 
  $$ \frac{\partial \sigma(x)}{\partial x}=\sigma(x)\cdot (1-\sigma(x))$$
  - loss function derivative: 
  $$\frac{\partial E}{\partial w_{kj}}= (t_k-o_k)\cdot (\sum_{i=1}^m w_{ki}h_i)\cdot (1-\sigma(\sum_{i=1}^m w_{ki}h_i))\frac{\partial}{\partial w_{kj}}\sum_{i=1}^m w_{ki}h_i$$
  $$\frac{\partial E}{\partial w_{kj}}=(t_k-o_k)\cdot (\sum_{i=1}^m w_{ki}h_i)\cdot (1-\sigma(\sum_{i=1}^m w_{ki}h_i))\cdot h_j$$
